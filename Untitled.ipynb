```python
import time
import pandas as pd
from bs4 import BeautifulSoup
import undetected_chromedriver as uc


cities = [
    "Delhi",
    "Mumbai",
    "Pune",
    "Kolkata",
    "Hyderabad",
    "Ahmedabad",
    "Bangalore",
    "Chennai"
]

BASE_URL = "https://www.oyorooms.com/hotels-in-{city}/"
SCROLL_COUNT = 10
MAX_PAGES = 3   


options = uc.ChromeOptions()
options.add_argument("--start-maximized")
driver = uc.Chrome(options=options)

results = []


def scroll_all():
    for _ in range(SCROLL_COUNT):
        driver.execute_script("window.scrollBy(0, 1200);")
        time.sleep(0.8)


for city in cities:
    print(f"\n===== CITY: {city} =====")

    for page in range(1, MAX_PAGES + 1):
        if page == 1:
            url = BASE_URL.format(city=city.lower())
        else:
            url = BASE_URL.format(city=city.lower()) + f"?page={page}"

        print(f"Scraping page {page}")
        driver.get(url)
        time.sleep(5)

        scroll_all()

        soup = BeautifulSoup(driver.page_source, "html.parser")

        hotel_cards = soup.find_all(
            "div", class_="hotelCardListing__descriptionWrapper"
        )

        print(f"Hotels found: {len(hotel_cards)}")

        if len(hotel_cards) == 0:
            print("No more pages. Moving to next city.")
            break

        for card in hotel_cards:
            try:
                hotel_name = card.find(
                    "h3", class_="listingHotelDescription__hotelName"
                )
                hotel_name = hotel_name.get_text(strip=True) if hotel_name else None

                address = card.find("span", itemprop="streetAddress")
                address = address.get_text(strip=True) if address else None

                rating = card.find("meta", itemprop="ratingValue")
                rating = rating["content"] if rating else None

                review_count = card.find("meta", itemprop="reviewCount")
                review_count = review_count["content"] if review_count else None

                price = card.find("span", class_="listingPrice__finalPrice")
                price = price.get_text(strip=True) if price else None

                results.append({
                    "city": city,
                    "hotel_name": hotel_name,
                    "address": address,
                    "rating": rating,
                    "review_count": review_count,
                    "price": price
                })

            except Exception as e:
                print("Error:", e)


df = pd.DataFrame(results)

df.to_csv("oyo_hotels_all_cities.csv", index=False)
df.to_excel("oyo_hotels_all_cities.xlsx", index=False)

print("\n✅ Saved files:")
print("✔ oyo_hotels_all_cities.csv")
print("✔ oyo_hotels_all_cities.xlsx")

driver.quit()
```
